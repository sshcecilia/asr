# Hot Words Detection

```python
import pandas as pd
```

```python
data = pd.read_csv('../data/cv-valid-dev.csv')
data.head()
```

## Detect Hot Words

Create a function that loops through each hot word and identify if it appears in text. If yes, return a list of hotwords that appears in text. Apply it to each row of text. For text with hot words detected, save its filename and list of hotwords into a text file.

```python
hotwords = ['be careful', 'destroy', 'stranger']

def detect_hotwords(text):
    text = text.lower()
    detected = []
    for i in hotwords:
        if i in text:
            detected.append(i)
    return detected
```

```python
data['hotwords'] = data['text'].apply(detect_hotwords)
```

```python
results = data[['filename', 'hotwords']][data['hotwords'].apply(len) != 0]
results.to_csv('detected.txt', sep='\t', index=False)
```
